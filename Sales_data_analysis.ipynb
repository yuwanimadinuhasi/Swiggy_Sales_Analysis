# Charts

### Monthly Sales Trend

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["YearMonth"] = df["Order Date"].dt.to_period("M").astype(str)

monthly_revenue = df.groupby("YearMonth")["Price (INR)"].sum().reset_index()

plt.figure()
plt.plot(monthly_revenue["YearMonth"], monthly_revenue["Price (INR)"])
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Revenue (INR)")
plt.title("Monthly Revenue Trend")
plt.tight_layout()
plt.show()
```

### Daily Sales Trend

```python
df["DayName"] = pd.to_datetime(df["Order Date"]).dt.day_name()

daily_revenue = (
    df.groupby("DayName")["Price (INR)"]
    .sum()
    .reindex(["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"])
)

plt.figure(figsize=(10,5))
plt.bar(daily_revenue.index, daily_revenue.values, color="skyblue")
plt.title("Daily Revenue Trend (Mon-Sun)")
plt.xlabel("Day")
plt.ylabel("Revenue (INR)")
plt.xticks(rotation=30)

plt.show()
```

### Total Sales by Food Type (Veg vs Non-Veg)

```python
non_veg_keywords = [
    "chicken", "egg", "fish", "mutton",
    "prawn", "biryani", "kabab", "kebab",
    "non-veg", "non veg"
]

df["Food Category"] = np.where(
    df["Dish Name"].str.lower().str.contains("|".join(non_veg_keywords), na=False),
    "Non-Veg",
    "Veg"
)
```

```python
food_revenue = (
    df.groupby("Food Category")["Price (INR)"]
    .sum()
    .reset_index()
)
```

```python
fig = px.pie(
    food_revenue,
    values="Price (INR)",
    names="Food Category",
    hole=0.5,
    title="Revenue Contribution: Veg vs Non-Veg",
    color_discrete_sequence=["#2ECC71", "#5DADE2"]
)

fig.update_traces(
    textinfo="percent+label",
    pull=[0.05, 0]
)

fig.update_layout(
    height=500,
    margin=dict(t=60, b=40, l=40, r=40)
)

fig.show()
```

### Total Sales by State

```python
fig = px.bar(
    df.groupby("State", as_index=False)["Price (INR)"].sum()
    .sort_values("Price (INR)", ascending=False),
    x="Price (INR)",
    y="State",
    orientation="h",
    title="Revenue by State (INR)"
)

fig.update_layout(height=600, yaxis=dict(autorange="reversed"))
fig.show()
```

### Quarterly Performance Summary

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Quarter"] = df["Order Date"].dt.to_period("Q").astype(str)

quarterly_summary = (
    df.groupby("Quarter", as_index=False)
    .agg(
        Total_Sales=("Price (INR)", "sum"),
        Avg_Rating=("Rating", "mean"),
        Total_Orders=("Order Date", "count")
    )
    .sort_values("Quarter")
)

quarterly_summary["Total_Sales"] = quarterly_summary["Total_Sales"].round(0)
quarterly_summary["Avg_Rating"] = quarterly_summary["Avg_Rating"].round(2)

quarterly_summary
```

### Top 5 Cities by Sales

```python
top_5_cities = (
    df.groupby("City")["Price (INR)"]
    .sum()
    .nlargest(5)
    .sort_values()
    .reset_index()
)

fig = px.bar(
    top_5_cities,
    x="Price (INR)",
    y="City",
    orientation="h",
    title="Top 5 Cities by Sales (INR)",
    color_discrete_sequence=["blue"]
)

fig.show()
```

### Weekly Trend Analysis

```python
df["Week"] = df["Order Date"].dt.isocalendar().week.astype(str)

weekly_revenue = (
    df.groupby("Week")["Price (INR)"]
    .sum()
    .reset_index()
    .sort_values("Week")
)

fig = px.line(
    weekly_revenue,
    x="Week",
    y="Price (INR)",
    title="Weekly Revenue Trend",
    markers=True
)

fig.update_layout(
    xaxis_title="Week Number",
    yaxis_title="Revenue (INR)",
    height=500
)

fig.show()
```

### Top 10 Most Popular Dishes

```python
top_dishes = df["Dish Name"].value_counts().head(10).reset_index()
top_dishes.columns = ["Dish Name", "Order Count"]

fig = px.bar(top_dishes, x="Order Count", y="Dish Name",
             orientation="h", title="Top 10 Most Ordered Dishes",
             color_discrete_sequence=["#f39c12"])
fig.update_layout(yaxis=dict(autorange="reversed"))
fig.show()
```

### Rating Distribution

```python
fig = px.histogram(df, x="Rating", nbins=10,
                   title="Distribution of Ratings",
                   color_discrete_sequence=["#3498db"])
fig.show()
```

### Average Order Value by City (Top 10)

```python
avg_city = (df.groupby("City")["Price (INR)"]
            .mean()
            .nlargest(10)
            .reset_index()
            .sort_values("Price (INR)"))

fig = px.bar(avg_city, x="Price (INR)", y="City",
             orientation="h",
             title="Top 10 Cities by Average Order Value",
             color_discrete_sequence=["#9b59b6"])
fig.show()
```
